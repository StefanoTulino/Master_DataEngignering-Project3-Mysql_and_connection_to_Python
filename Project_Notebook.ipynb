```python
## [FIRST PROJECT ON PYTHON AND MYSQL]

import mysql.connector
## Punto 1: Connessione ad un db MySQL

mycon = mysql.connector.connect(
    host='localhost', user='root',
    password='root',
    database='banca')

mycur = mycon.cursor()


## Punto 2: Leggere in banca dati per visualizzare solo gli utenti con almeno 100 transazioni in uscita
def first_select_clients():
    sql= """
    SELECT distinct c.id_cliente,c.nome,c.cognome, count(t.data) as num_transazioni_uscita,
        count( case when t.id_tipo_trans=3 then t.id_tipo_trans else null end) as 'Acquisto su Amazon',
        count( case when t.id_tipo_trans=4 then t.id_tipo_trans else null end) as 'Rata Mutuo',
        count( case when t.id_tipo_trans=5 then t.id_tipo_trans else null end) as 'Hotel',
		count( case when t.id_tipo_trans=6 then t.id_tipo_trans else null end) as 'Biglietto Aereo',
		count( case when t.id_tipo_trans=7 then t.id_tipo_trans else null end) as 'Supermercato'
    FROM banca.cliente c
        left join banca.conto cont on c.id_cliente=cont.id_cliente
            left join banca.transazioni t on cont.id_conto=t.id_conto
                left join (
                        select *
                            from banca.tipo_transazione tipo_t
                        where tipo_t.segno='-') usc
                on t.id_tipo_trans=usc.id_tipo_transazione
    WHERE (usc.segno is not null and cont.id_conto is not null) or (usc.segno is null and cont.id_conto is null)
    group by 1,2,3
    HAVING num_transazioni_uscita>100
    order by c.id_cliente;
    """

    mycur.execute(sql)    

    d = mycur.fetchall()
    list_of_ids = []
    for ids in d:
        list_of_ids.append(ids)
         
    return list_of_ids


## Punto 3) Leggere in banca dati per visualizzare solo gli utenti con almeno 20 transazioni in entrata
def second_select_clients():
    sql= """
    SELECT distinct c.id_cliente,c.nome,c.cognome, count(t.data) as num_transazioni_entrata,
        count( case when t.id_tipo_trans=0 then t.id_tipo_trans else null end) as 'Stipendio',
		count( case when t.id_tipo_trans=1 then t.id_tipo_trans else null end) as 'Pensione',
		count( case when t.id_tipo_trans=2 then t.id_tipo_trans else null end) as 'Dividendi'
    FROM banca.cliente c
        left join banca.conto cont on c.id_cliente=cont.id_cliente
            left join banca.transazioni t on cont.id_conto=t.id_conto
                left join (
                        select *
                            from banca.tipo_transazione tipo_t
                        where tipo_t.segno='+') entr
                on t.id_tipo_trans=entr.id_tipo_transazione
    WHERE (entr.segno is not null and cont.id_conto is not null) or (entr.segno is null and cont.id_conto is null)
    group by 1,2,3
    HAVING num_transazioni_entrata>20
    order by c.id_cliente;
    """

    mycur.execute(sql)    
    d = mycur.fetchall()    
    return d

   
# MAIN ------------------------------------

list1=first_select_clients()
print("First Query= ",list1,"\n")

list2=second_select_clients()
print("Second Query= ",list2,"\n")
```
